```python
from utility.utilities import read_excel_sheet, adjust_team_cells

import os
```

```python
src_dir = os.path.abspath(os.path.join(os.getcwd(), "."))

file_absolute_path_excel = os.path.join(src_dir, "data/NBA23-24.xlsx")

df_average = read_excel_sheet(file_absolute_path_excel, "StatsPerGame")
```

```python
df_average.head(15)
```

```python
df_average.columns
```

```python
no_columns = ['GS', '3PA','3P%','2P','2PA','2P%','eFG%', 'ORB', 'DRB','PF','Awards']
df_average.drop(no_columns, axis=1, inplace=True)
```

```python
df_average.head(15)
```

```python
print("Missing values:")
print(df_average.isnull().sum())
df_average[df_average.isnull().any(axis=1)].head(10)
```

**The NULL Values are only percentage values for a few players. Therefore we can fill the cells with 0.0**

```python
df_average = df_average.fillna(0.0)
```

```python
Multiple_teams=df_average[(df_average["Team"] == '2TM') | (df_average["Team"] == '3TM')]
Multiple_teams.tail(20)
```

```python
# Step 1: Get a list of index values where "Team" is '2TM' or '3TM'
target_indexes = df_average.index[df_average['Team'].isin(['2TM', '3TM'])].tolist()

print(target_indexes)

# Step 2: Update each combined row and drop individual rows
for target_index in target_indexes:
    print(target_index)
    df_average = adjust_team_cells(df_average, target_index)
df_average.reset_index(drop=True, inplace=True)
df_average.set_index('Rk', inplace=True)
```
